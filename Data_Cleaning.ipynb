# Defining Folder Paths

```python
OG_Data_Folder = './OG'
OG_Train = os.path.join(OG_Data_Folder, 'train')
OG_Test = os.path.join(OG_Data_Folder, 'test')
OG_Val = os.path.join(OG_Data_Folder, 'val')
```

# Copying the Original Dataset into a dummy one

```python
New_Folder_Name = "./V1"

shutil.copytree(OG_Data_Folder, New_Folder_Name)
```

# Defining the subset labels, which are more or less equal in their number of instances

```python
filter_list = [
    'Bottle cap',
    'Bottle',
    'Can',
    'Carton',
    'Cup',
    'Paper',
    'Straw',
    'Unlabeled litter',
]

filter_list
```

```python
Data_Classes = ['Aluminium foil', 'Bottle cap', 'Bottle', 'Broken glass', 'Can', 'Carton', 'Cigarette', 'Cup', 'Lid', 'Other litter', 'Other plastic', 'Paper', 'Plastic bag - wrapper', 'Plastic container', 'Pop tab', 'Straw', 'Styrofoam piece', 'Unlabeled litter']
Actual_label2idx = {label: idx for idx, label in enumerate(Data_Classes)}
Actual_idx2label = {idx: label for idx, label in enumerate(Data_Classes)}

Actual_label2idx, Actual_idx2label
```

# Removing the unwarnted labels from each image's label file, so yolo doesn't learn those labels.

```python
for idx_mode, mode in enumerate(["train", "valid", "test"]):
    path_images = os.path.join(New_Folder_Name, mode, "images")
    path_labels = os.path.join(New_Folder_Name, mode, "labels")

    for image_file, label_file in zip(os.listdir(path_images), os.listdir(path_labels)):
        label_file_content = open(os.path.join(path_labels, label_file), "r").readlines()
        copy_of_file_content = label_file_content.copy()
        check = False
        for line in label_file_content:
            label = line.split(" ")[0]
            if Actual_idx2label[int(label)] not in filter_list:
                copy_of_file_content.remove(line)
                check = True
        open(os.path.join(path_labels, label_file), "w").writelines(copy_of_file_content)

        if check == True:
            print(f'File changed at path: {os.path.join(path_labels, label_file)}')
```
